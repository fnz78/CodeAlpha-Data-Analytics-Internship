```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (8,5)
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("cleaned_books_data.csv")

df.head()
```

```python
df.info()
```

```python
sns.histplot(df["Price"], bins=8)
plt.title("Distribution of Book Prices")
plt.xlabel("Price (Â£)")
plt.ylabel("Number of Books")
plt.show()
```

### Insight:
Most books are concentrated in the mid-price range.
There are very few extreme low or high prices.
This indicates price clustering.

```python
sns.countplot(x="Rating", data=df)
plt.title("Book Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

### Insight:
Rating 3 appears most frequently.
Very few books have 1-star ratings.
This suggests moderate overall quality.

```python
sns.boxplot(x="Rating", y="Price", data=df)
plt.title("Price Variation by Rating")
plt.show()
```

### Insight:
Prices overlap across ratings.
Higher rating does not strongly mean higher price.

```python
avg_price = df.groupby("Rating")["Price"].mean()

avg_price.plot(kind="bar")
plt.title("Average Price by Rating")
plt.xlabel("Rating")
plt.ylabel("Average Price (Â£)")
plt.show()
```

### Insight:
The average price does not increase significantly with rating.
This indicates weak relationship between price and quality.

```python
top_expensive = df.sort_values(by="Price", ascending=False).head(5)

sns.barplot(x="Price", y="Title", data=top_expensive)
plt.title("Top 5 Most Expensive Books")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
